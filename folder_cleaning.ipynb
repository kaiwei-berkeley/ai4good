```python
import os
import re
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_dir = ".\data\\0001"
all_file = os.listdir(data_dir)
```

```python
all_folders = [i for i in all_file if re.search("\.pdf",i)==None]

#folder_dir = [data_dir+"\\"+i for i in all_folders]

lists_of_file = {i:os.listdir(data_dir+"\\"+i) for i in all_folders}
```

```python
#get extension
print(all_folders[0])
[i.split(".")[-1] for i in lists_of_file[all_folders[1]]]

# get size
[os.path.getsize(data_dir+"\\"+all_folders[0]+"\\"+i) for i in lists_of_file[all_folders[0]]]

# get time
pd.to_datetime([os.path.getmtime(data_dir+"\\"+all_folders[0]+"\\"+i) for i in lists_of_file[all_folders[0]]],unit='s')
```

```python
# get all extension
extension_list = [[i.split(".")[-1].lower() for i in lists_of_file[folder]] for folder in all_folders]
all_extension = [item for sublist in extension_list for item in sublist] 

## get most common extension
print('unique extension:',len(pd.unique(all_extension)))
pd.unique(all_extension)
count = Counter()
for i in pd.unique(all_extension):
    count[i] = all_extension.count(i)
count.most_common(10)
```

```python
eps_count = [i.count('eps') for i in extension_list]
ps_count = [i.count('ps') for i in extension_list]
tex_count = [i.count('tex')+i.count('ltx') for i in extension_list]
all_count = [len(i) for i in extension_list]

## distribution of number of tex
count = Counter()
for i in tex_count:
    count[i] = tex_count.count(i)
count
```

```python
last_mtime = [pd.to_datetime(max([os.path.getmtime(data_dir+"\\"+folder+"\\"+i) for i in lists_of_file[folder]]),unit='s') 
              for folder in all_folders]
```

```python
tex_path = []
for folder in lists_of_file:
    r = re.compile('.*\.tex|.*\.ltx|.*\.TEX')
    tex_files = list(filter(r.match, lists_of_file[folder]))
    if len(tex_files) == 0:
        tex_path.append("")
    else:
        ## find largest size .tex
        index_max = np.argmax([os.path.getsize(data_dir+"\\"+folder+"\\"+i) for i in tex_files])
        tex_path.append(data_dir+"\\"+folder+"\\"+tex_files[index_max])
tex_path
```

```python
dat = pd.DataFrame({'all_folders':all_folders,'eps_count':eps_count,'ps_count':ps_count,
                   'tex_count':tex_count,'tex_path':tex_path,'last_mtime':last_mtime,
                   'all_count':all_count})
dat.to_csv(".\data\info.csv",index=False)
```

```python
import re
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dat = pd.read_csv(".\data\info.csv",keep_default_na = False)
```

```python
print(dat['tex_path'])
```

```python
abstract = []
abst_count = []
for i,path in enumerate(dat['tex_path']):
    print (i)
    print(path)
    if path != '':
        f = open(path,'r',encoding="utf8", errors='ignore')
        lines = f.readlines()
        f.close
        index = [i for i,word in enumerate(lines) if re.search('\{abstract\}',word)]
        if len(index) <= 1:
            abst_count.append(0)
            abstract.append("")
        else:
            sens = lines[(index[0]+1):index[1]]
            abst_count.append(len(index))
            abstract.append(" ".join(sens))
    else:
        abst_count.append(0)
        abstract.append("")
```

```python
word_count = [ab.count(" ") for ab in abstract]
```

```python
sns.distplot(word_count)
word_count_dist = Counter(word_count)
word_count_dist.most_common(20)
```
